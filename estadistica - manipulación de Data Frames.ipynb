# Manipulacion de DataFrames

```python
df3 = pd.DataFrame({
    'key' : ['A', 'B', 'C', 'A', 'B',  'C'],
    'value1': range(6),
    'value2': np.random.randint(0,100,6)
})
```

```python
df3
```

# Aggregate 

```python
df3.groupby('key').aggregate(['min', np.median, np.mean, max])
```

```python
df3.groupby('key').aggregate({
    'value1': 'min',
    'value2': 'max'    
})
```

# Filter 

```python
def filter_func(x):
    return x['value2'].std()>5
```

```python
df3.groupby('key').filter(filter_func)
```

```python
df3.groupby('key').transform(lambda x: x - x.mean())
```

## Apply

```python
def norm_by_col2(x):
    x['value1'] /= x['value2'].sum()  # o lo que es igual: x['value1']= x['value1']/x['value2'].sum()
    return x
```

```python
df3.groupby('key').apply(norm_by_col2)
```

# Gr√°ficos desde data frames y exportaci√≥n

```python
import numpy as np
years = [year for year in range(1900,2020)]
deads = [(y + np.random.uniform(0,1000)- 1850) for y in years]
```

```python
df = pd.DataFrame({
    "year": years,
    "deads": deads 
    
})
```

```python
my_plot = df.plot(x='year', y = 'deads', figsize = (10,8))
```

```python
my_fig = my_plot.get_figure()
```

```python
my_fig.savefig('gr√°fico.png')
```
